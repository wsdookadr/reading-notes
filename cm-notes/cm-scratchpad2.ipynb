**Problem 2.22**

Solution, due to Joel:

$\mathrm{RULE}(a_i, b_i):=$
\begin{align}
\left(\sum_{1\leq i\leq n} a_i\right)\left(\sum_{1\leq i\leq n} b_i\right)
&= \sum_{1\leq i\leq n} \sum_{1\leq j\leq n} a_i b_j \\
&= \sum_{\substack{1\leq i\leq n\\1\leq j\leq n}} a_i b_j \\
&= \sum_{i\lt j\;\mathrm{or}\;i=j\;\mathrm{or}\;i\gt j} a_i b_j \\
&= \sum_{i\lt j} a_i b_j + \sum_{i=j} a_i b_j + \sum_{i\gt j} a_i b_j \\
&= \sum_{i} a_i b_i + \sum_{i\gt j} (a_i b_j + a_j b_i) \\
\end{align}

$\mathrm{RULE}(a_i b_i, a_i b_i)$



We start from the RHS, which we can rewrite using the rules above as:  $\mathrm{RULE}(a_i^2, b_i^2) - \mathrm{RULE}(a_i b_i, a_i b_i)$ . If we expand accordingly, the $\sum a_i^2 b_i^2$ term will cancel out and we can regroup the terms in a way that looks like LHS.

Note: also check the [related blog post here](https://geq-i.blogspot.com/2015/01/cauchyschwarz.html) shows how those sums of squares will cancel out.

For the 2nd part of the problem, the rule should be: $\mathrm{RULE}(a_i A_i, b_i B_i) - \mathrm{RULE}(a_i B_i,A_i b_i)$

Talked to Joel about this, he mentioned Wildberger's [six playlist](https://www.youtube.com/playlist?list=PLzdiPTrEWyz4KD007Ge10dfrDVc4YwlYS) in relation to finding the rule here.



$\mathrm{RULE}(a_i a_i, b_i b_i) - \mathrm{RULE}(a_i b_i, a_i b_i)$

$\sum_{1\leq i\lt j\leq n}(a_i b_j - a_j b_i)(A_i B_j - A_j B_i)$

$\mathrm{RULE}(a_i A_i, b_i B_i) - \mathrm{RULE}(a_i B_i, A_i b_i):$
\begin{align}
0 &= \left(\left(\sum_{1\leq i\leq n} a_i A_i\right)\left(\sum_{1\leq i\leq n} b_i B_i\right)
 - (\sum_{i} a_i A_i b_i B_i + \sum_{i\lt j} (a_i A_i b_j B_j + a_j A_j b_i B_i))\right) \\
  &\quad- \left(\left(\sum_{1\leq i\leq n} a_i B_i\right)\left(\sum_{1\leq i\leq n} A_i b_i\right)
 - (\sum_{i} a_i B_i A_i b_i + \sum_{i\lt j} (a_i B_i A_j b_j + a_j B_j A_i b_i))\right) \\
  &= \left(\sum_{1\leq i\leq n} a_i A_i\right)\left(\sum_{1\leq i\leq n} b_i B_i\right) - \left(\sum_{1\leq i\leq n} a_i B_i\right)\left(\sum_{1\leq i\leq n} A_i b_i\right) + \sum_{i\lt j} ((a_i A_i b_j B_j + a_j A_j b_i B_i) - (a_i B_i A_j b_j + a_j B_j A_i b_i)) \\
  &= \left(\sum_{1\leq i\leq n} a_i A_i\right)\left(\sum_{1\leq i\leq n} b_i B_i\right) - \left(\sum_{1\leq i\leq n} a_i B_i\right)\left(\sum_{1\leq i\leq n} A_i b_i\right) + \sum_{i\lt j} (a_i b_j - a_j b_i)(A_i B_j - A_j B_i)
\end{align}

Partitions $\{a, A, b, B\} \leadsto$
\begin{align}
  &\{\{a, A\}, \{b, B\}\} \tag{P1} \\
  &\{\{a, b\}, \{A, B\}\} \tag{P2} \\
  &\{\{a, B\}, \{A, b\}\} \tag{P3}
\end{align}

$\mathrm{RULE}(\mathrm{P1}) - \mathrm{RULE}(\mathrm{P3}) = \dots + \sum_{i\lt j}\mathrm{P2}$

$\mathrm{RULE}(\mathrm{P1}) - \mathrm{RULE}(\mathrm{P3}):$
$$0 = [\sum\sum]\mathrm{P1} - [\sum\sum]\mathrm{P3} + \sum_{i\lt j}\mathrm{P2}$$


**Problem 2.11**

$E$ is the shift operator.

$$
\begin{align}
  & \hskip 4pt \sum \Delta a_k b_k & \tag{1} \\
= & \sum_{0\leq k < n} (a_{k+1} - a_k)b_k  & \tag{2}\\
= & \sum_{0\leq k < n} a_{k+1} b_k - a_k b_k  & \tag{2}\\
= & \sum_{0\leq k < n} a_{k+1} b_k - \sum_{0\leq k < n} a_k b_k  & \tag{3}\\
= & \sum_{0\leq k < n} a_{k+1} b_k - \sum_{0\leq k+1 < n} a_{k+1} b_{k+1}  & \tag{4}\\
= & \sum_{0\leq k < n} a_{k+1} b_k - \sum_{-1\leq k < n-1} a_{k+1} b_{k+1}  & \tag{5}\\
= & \sum_{0\leq k < n} a_{k+1} b_k - \sum_{0\leq k < n} a_{k+1} b_{k+1} + (a_n b_n - a_0 b_0)  & \tag{6}\\
= & - \sum_{0\leq k < n} a_{k+1} (b_{k+1} - b_k) + (a_n b_n - a_0 b_0)  & \tag{7}\\
= & - \sum_{0\leq k < n} a_{k+1} \Delta b_{k} + \left.(a_k b_k)\right|_0^n   & \tag{8}\\
= & \left.(a_k b_k)\right|_0^n - \sum_{0\leq k < n} E a_k \Delta b_{k}  & \tag{9}\\
\end{align}
$$

Step (5) will shift $k \to k+1$.

Step (6) will re-adjust the bounds. 

Step (7) splits off the first term, adds an addition term, and then subtracts that additional term. 

Step (9) rewrites the outer difference as an evaluation, also rewrites the sum using the difference operator.

Step (10) uses rewrites the expression using the shift operator.

**Note:** the evaluation in (10) has its upper bound 1 greater than the initial upper-bound of the original sum in (1).

\end{align*}
