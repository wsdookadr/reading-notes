Split off the first term:

$U_{n+1} = (n+1)^2 + \sum_{k=1}^{n+1} (-1)^{k} (n+1-k)^2$

$$
\begin{array}{|c|c|}
\hline
\textbf{shift} & \textbf{general term} & \textbf{bound} \downarrow & \textbf{bound} \uparrow \\
\hline
0 & (-1)^k (n+1-k)^2 & 1 & n+1 \\
\hline
k\to k+1 & (-1)^{k+1} (n-k)^2 & 0 & n \\
\hline
k\to k-1 & (-1)^{k-1} (n+2-k)^2 & 2 & n+2 \\
\hline
\end{array}
$$

Using the shift $k \to k+1$ :

$$ 
U_{n+1} = (n+1)^2 + \sum_{k=0}^{n} (-1)^{k+1} (n-k)^2 = (n+1)^2 - \sum_{k=0}^{n} (-1)^{k} (n-k)^2 = (n+1)^2 - U_n
$$

$$
\begin{align}
U_{n+1} & = 0 + \sum_{k=0}^{n} (-1)^{k} (n+1-k)^2 \\
