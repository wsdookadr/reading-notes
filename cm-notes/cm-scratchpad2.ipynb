Solution, due to Joel:

$\mathrm{RULE}(a_i, b_i):=$
\begin{align}
\left(\sum_{1\leq i\leq n} a_i\right)\left(\sum_{1\leq i\leq n} b_i\right)
&= \sum_{1\leq i\leq n} \sum_{1\leq j\leq n} a_i b_j \\
&= \sum_{\substack{1\leq i\leq n\\1\leq j\leq n}} a_i b_j \\
&= \sum_{i\lt j\;\mathrm{or}\;i=j\;\mathrm{or}\;i\gt j} a_i b_j \\
&= \sum_{i\lt j} a_i b_j + \sum_{i=j} a_i b_j + \sum_{i\gt j} a_i b_j \\
&= \sum_{i} a_i b_i + \sum_{i\gt j} (a_i b_j + a_j b_i) \\
\end{align}

$\mathrm{RULE}(a_i b_i, a_i b_i)$



We start from the RHS, which we can rewrite using the rules above as:  $\mathrm{RULE}(a_i^2, b_i^2) - \mathrm{RULE}(a_i b_i, a_i b_i)$ . If we expand accordingly, the $\sum a_i^2 b_i^2$ term will cancel out and we can regroup the terms in a way that looks like LHS.

Note: also check the [related blog post here](https://geq-i.blogspot.com/2015/01/cauchyschwarz.html) shows how those sums of squares will cancel out.

By symmetry, the indices can be renamed $P=\prod_{1\leq j \leq k \leq n} a_j a_k=\prod_{1\leq k \leq j \leq n} a_j a_k$

Converting the Iverson bracket identity

$ [1\leq j\leq k\leq n] + [1\leq k\leq j \leq n] = [1\leq j,k \leq n] + [1\leq j=k \leq n] $

to products, yields:

\begin{array}{rll}
\prod_{1\leq j \leq k \leq n} a_j a_k \cdot \prod_{1\leq k \leq j \leq n} a_j a_k & = &
 \prod_{1\leq j,k \leq n} a_j a_k \cdot \prod_{1\leq k=j \leq n} a_j a_k \\
P^2 & = & \underbrace{\prod_{1\leq j,k \leq n} a_j a_k}_{A} \cdot \underbrace{\prod_{1\leq k=j \leq n} a_j a_k}_{B} \\
\end{array}

\begin{array}{ll}
A & = \prod_{1\leq j,k \leq n} a_j a_k \\
  & = \prod_{1\leq j\leq n} \prod_{1\leq k\leq n} a_j a_k\\

\end{array}

\begin{array}{ll}
